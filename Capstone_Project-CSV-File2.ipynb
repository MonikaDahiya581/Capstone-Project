```python
# importing the required libraries
from bs4 import BeautifulSoup,SoupStrainer
import requests
import pandas as pd
import numpy as np
```

```python
#Send get request to the webpage server to get the source code of the page
url="https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
IMBD = requests.get(url)
soup = BeautifulSoup(IMBD.content, 'html.parser')
```

```python
movie_name =[] 
votes =[] 
Genre =[] 
gross_collection =[] 
popularity = [] 
Certification =[]
```

```python
movie_data = soup.findAll('div',attrs = {'class': 'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
```

```python
print(len(movie_name))
```

```python
#creating empty list for stars
starList = list()
for star in movie_data:
    starList.append(star.find("p",attrs = {"class": ""}).text.replace("Stars:", "").replace("\n", "").strip())
    
    
    
for store in movie_data:    
 
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote = value[0].text
    votes.append(vote)
    
    grosses = value[1].text if len(value) >1 else '******'
    gross_collection.append(grosses)
    
    genre = store.p.find('span', class_='genre')
    Genre.append(genre)
    
   
    popular = store.find('div', class_= 'inline-block ratings-imdb-rating').text.replace('\n', '')
    popularity.append(popular)
    
    certificate = store.p.find('span', class_ ='certificate')
    Certification.append(certificate)
    
    
```

```python
df = pd.DataFrame({'Movie_name':movie_name,'Stars':starList,'Votes':votes,'Genres':Genre,'Gross_collection':gross_collection,'Popularity':popularity,'Certificates':Certification})
```

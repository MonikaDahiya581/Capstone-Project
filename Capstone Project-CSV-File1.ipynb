```python
# importing the required libraries
from bs4 import BeautifulSoup,SoupStrainer
import requests
import pandas as pd
import numpy as np
```

```python
#Send get request to the webpage server to get the source code of the page
url="https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
IMBD = requests.get(url)
soup = BeautifulSoup(IMBD.content, 'html.parser')
```

# Using Beautiful Soup Extract Information from the IMDB web page


```python
#creating the empty list
sno = []

movie_data = soup.findAll('div',attrs = {'class': 'lister-item mode-advanced'})
```

```python
# Firstly Get The Serial No for Movies Data
for store in movie_data:
    serial = store.h3.find('span', class_='lister-item-index unbold text-primary').text.replace('(','').replace(')','')
    sno.append(serial)
```

```python
#First Of All we Get the movie_data
movie_data = soup.findAll('div',attrs = {'class': 'lister-item mode-advanced'})

#Get The Movie_name
def movie_name(soup):
    
    movie_class="lister-item-header"
    movie_name_tags=soup.find_all('h3',{'class':movie_class})
    movie_name=[]

    for tag in movie_name_tags:
        moviename = tag.find('a').text
        movie_name.append(moviename) 
        
    return movie_name
```

```python
moviename = movie_name(soup)
```

```python
print(len(moviename))
```

```python
moviename[:10]
```

# Directors Name

```python
def director_name(soup):
    
    director_selection = ""
    movie_director_tags=soup.find_all('p',{'class':director_selection})
    movie_director=[]
    
    for tag in movie_director_tags:
        director = tag.find('a').text
        movie_director.append(director)
    
    return movie_director
```

```python
director = director_name(soup)
```

```python
director[:10]
```

```python
print(len(director))
```

# Duration:Runtime

```python
duration = []
```

```python
for store in movie_data:
    duration_time = store.p.find('span', class_='runtime').text
    duration.append(duration_time)
     
```

# Year Of Release

```python
year = []
```

```python
for store in movie_data:
    year_of_release = store.h3.find('span',class_= 'lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_release)
      
```

# Ratings

# SQLite3 Queries Through Pandas DataFrame

```python
from pandasql import sqldf
```

```python
con=sqlite3.connect('myimdbTable.db')
```

```python
imdb.to_sql(name="imdb_movies",con=con,if_exists='replace',index=False)
```

```python
con.commit()
```

```python
imdb_df = pd.read_sql_query('select * from imdb_movies', con)
```

```python
imdb_df
```

# Q.1)  Display all the details of movies created by directors Christopher and Matt Reeves.

```python
#select the Directors names from movies

query1 = "SELECT * FROM imdb_df Where Director_1 == 'Christopher Nolan' OR Director_2 == 'Matt Reeves'"
name_directors = sqldf(query1)
name_directors
```

```python
query11 = "SELECT * FROM imdb_df WHERE Director_1 BETWEEN 'Christopher Nolan' AND 'Matt reeves' "
sqldf(query11,globals())
```
